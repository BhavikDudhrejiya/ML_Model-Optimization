# Defining the parameter for Gredient Descent

```python
#Equation for linear regression is y = mx + c
#m is a slope and c is a intercept

#Defining the parameter
m = 0 #Slope
c = 0 #Intercept

#Definning learning rate
L = 0.0001

#Definning iterations require
epochs = 1000

#Defining number of elements
n = float(len(X))
```

# Implementing Gredient Descent

```python
#Implementing gredient descent
for i in range(epochs):
    y_pred = m*X+c
    D_m = -(2/n)*sum(X*(y - y_pred))
    D_c = -(2/n)*sum(y - y_pred)
    m = m - L*D_m
    c = c - L*D_c
    cost = (1/n) * sum([val**2 for val in (y-y_pred)])
    print('m = {}, c = {}, cost {}, iteration = {}'.format(m,c,cost,i))
```
