# Basic Analysis of the data

```python
#Importing library
import numpy as np
import pandas as pd
```

```python
#Loading the data
dataset = pd.read_csv('C:\\Users\\bhavi\\OneDrive\\Desktop\\Data\\Iris.csv')
```

```python
#Checking the data
dataset.head()
```

# Splitting data into X & y

```python
#Splitting data into X & Y
X = dataset.iloc[:, 0:5].values #Selecting all the rows and 0 to 4 columns
y = dataset.iloc[:, 5].values #Selecting all the rows and 4th column
```

```python
#Checking the data
print(X[0:5])
print(y[0:5])
```

# Splitting data into Train and Test

```python
#Importing library for train, test & split
from sklearn.model_selection import train_test_split

#Splitting data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Feature Scaler of the data

```python
#Importing library for feature scaling
from sklearn.preprocessing import StandardScaler

#Implementing feature scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

#View the data
X_train
```

# Linear Discriminant Analysis

```python
#Importing library for linear disciminat analysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

#Implementing LDA
lda = LDA(n_components=1)
X_train = lda.fit_transform(X_train, y_train)
X_test = lda.transform(X_test)
```

# Random Forest Classifier

```python
#Imorting library for random forest
from sklearn.ensemble import RandomForestClassifier

#Implementing Random Forest
classifier = RandomForestClassifier(max_depth=2, random_state=0)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

# Confusion Matrix and Accuracy

```python
#Importing libraries for confusion matrix and accuracy
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

#Creating confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

#Getting accuracy
print('Accuracy' + str(accuracy_score(y_test, y_pred)))
```
